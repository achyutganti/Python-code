```python
import bs4 as bs
import lxml
from urllib import urlopen
import pandas as pd
```

```python
# Open a url by using this library function
sauce = urlopen('https://www.macys.com/shop/mens-clothing/mens-casual-shoes?id=59851&cm_sp=us_hdr-_-men-_-59851_casual-shoes_COL2').read() # Don,t forget .read()
# The above code just outputs the source code of the page which has all the 
# HTML code in it. So let us create a beautiful soup object to view it 
# properly.

print(sauce)
```

```python
# Creating a beautiful soup object
soup = bs.BeautifulSoup(sauce,'lxml') # Have to mention the parser, else, you get the warning. 
# Different machines use diffrnt parsers.

print(soup)

# The main difference between the sauce and the soup object is that 
# both sauce and soup objects return the source code as it is. But 
# soup's code is much readable and we can actually call soup.title and 
# other things on it and we cannot do that on the sauce code. 
```

```python
print(soup.title.text)
# you can even do, soup.title.name, soup.title.text, soup.title.string to 
# get the string version of that output. etc.  
```

```python
# print the first paragraph element of the page
print(soup.p)
```

```python
# if you want to print al the paragraphs at once, we can do this. 
print(soup.find_all('p'))

# We dont want the tags along with the text. so we can loop it
#for para in soup.find_all('p'):
print(para.text) 
# we can even do, para.string but that works only if
# you dont have any child tags in the source code. Else you get NOne.
```

```python
# if we want the whole document to print or access, we can do something like this.
print(soup.get_text())
```

```python
# to find all of the links <url> in the page,

for url in soup.find_all('a'):
    print(url.get('href'))
```

```python
# the above code gives you the links in the body, but what if you want to 
# get all the links in the navigation bar above. 
nav = soup.nav

for url in nav.find_all('a'):
    print(url.get('href'))
```

```python
body= soup.body
for para in body.find_all('p'):
    print(para.text)
```

```python
# sometimes we want to parse the div section. For that

for div in soup.find_all('div',class_='body'):
    print(div.text)  # Does not always give you the text. 
    # There has to exist text in between the div tags. 
# ?????Have to find out about this one????????
```

```python
# We will scrape the table information from the page

table = soup.table
# We can even do table = soup.find('table')
print(table)
```

```python
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row =[i.text for i in td]
    print(row)
```

```python
# Let us see how to do some dynamic Javascript scrapping in Python
import sys
from PyQt5.QtWidgets import QApplication # Different from the gt4 version
from PyQt5.QtWebKitWidgets import QWebView,QWebPage
from PyQt5.QtCore import QUrl
```

```python
class Client(QWebPage):
    def __init__(self,url):
        self.app = QApplication(sys.argv)
        QWebPage.__init__(self)
        self.loadFinished.connect(self.pageload)
        self.mainFrame().load(QUrl(url))
        self.app.exec_()
        
    def pageload(self):
        self.app.quit()

url = 'https://pythonprogramming.net/parsememcparseface/'
client_response = Client(url)
source = client_response.mainFrame().toHtml()
soup = bs.BeautifulSoup(source,'lxml')
js_text = soup.find('p',class_='jstest')
print(js_text.text)
```
