```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Creating the model with the keras library. 

# You can create a sequential model by passing a list of layer instances to the constructor. 

model = Sequential()
model.add(Dense(32, activation='relu',input_shape=(500,)))
model.add(Dense(10))
model.summary()
```

```python
#Next comes the compilation

model.compile(optimizer='rmsprop',
             loss = 'binary_crossentropy',
             metrics= ['accuracy'])
```

```python
# NExt step is to train the model 

model = Sequential()
model.add(Dense(32, activation = 'relu', input_dim = 100))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
             loss = 'binary_crossentropy',
             metrics= ['accuracy'])

```

```python
# Generate the data (dummy)

import numpy as np
data = np.random.random((1000,100))
labels = np.random.randint(2,size = (1000,1))
```

```python
model.summary()
```

```python
# Fitting the model

model.fit(data,labels,batch_size = 32, epochs = 10, verbose=2,
         validation_split=0.2,
         shuffle =True)
```

```python
# Evaluate the model. 

model.evaluate(data,labels,batch_size = 32, 
               verbose = 1, sample_weight = None)
```

```python
# You can predict the new data labels
model.predict(data,batch_size = 32,verbose =1)
```

```python
# We can predict classes. 

model.predict_classes(data,batch_size = 32,verbose =1)
```

```python
# Ypu can predict the prob which is the same as the first one 

model.predict_proba(data,batch_size = 32, verbose = 1)
```
